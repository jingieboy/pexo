# PSR J0740+6620
<a id='top'></a>

This demo compares the Shapiro delay in PSR J0740+6620, observed by Cromarti et al. 2019, using the orbital parameters from the paper.

## Parameters:

`-c T` - compute the timing model only,

`-t '2456640.5 2458462.5 0.5'` - model a list of JDs from 2456640.5 to 2458462.5 with a step of 0.5 day,

`-p ../input/PSR_J0740+6620.par` - [system parameters file](../input/PSR_J0740+6620.par), containing the orbital parameters of the system,

`-o ../results/PSRJ0740+6620.txt` - output file path,

`-v "BJDtdb1 BJDtdb2 ShapiroTarget"` - parameters to write to the output file - BJDTDB and Shapiro delay in the target system.

## Run the code

Execute PEXO with the command below. This may take a few minutes to compute.

```python
# cd into the code directory
%cd ../code
# run PEXO via the R script
! Rscript pexo.R -c T -t "2456640.5 2458462.5 0.5" -p ../input/PSR_J0740+6620.par -o ../results/PSRJ0740+6620.txt -v "BJDtdb1 BJDtdb2 ShapiroTarget"
```
