In this demo we simulate the α Centauri system.

## Parameters:

`-m emulate` - emulation mode,

`-c TA` - compute timing and astrometry models,

`-t ../input/gaia80yrby10day.tim` - [timing file](../input/gaia80yrby10day.tim), containing MJDs from 42000 to 52000 with a step of 10 days,

`-p ../input/ACAgaia.par` - [system parameters file](../input/ACAgaia.par), containing the information about the stellar system configuration.

## Run the code

Execute PEXO with the command below. This may take a few minutes to compute.

```python
# cd into the code directory
%cd ../code
# run PEXO via the R script
! Rscript pexo.R -m emulate -c TA -t ../input/gaia80yrby10day.tim -p ../input/ACAgaia.par
```
